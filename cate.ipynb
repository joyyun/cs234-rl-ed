```python
# create dataframe with only the rows where arm_name is 'Control'
control_df = summary_df[summary_df['arm_name'] == 'Control']

# create dataframe with only the rows where arm_name is 'Tutor Feedback'
tutor_df = summary_df[summary_df['arm_name'] == 'Tutor Feedback']

# create dataframe with only the rows where arm_name is 'Tutor Feedback + Goal-Oriented Learner Feedback'
tutor_learner_goal_df = summary_df[summary_df['arm_name'] == 'Tutor Feedback + Goal-Oriented Learner Feedback']

# create dataframe with only the rows where arm_name is 'Tutor Feedback + Socially-Oriented Learner Feedback'
tutor_learner_social_df = summary_df[summary_df['arm_name'] == 'Tutor Feedback + Socially-Oriented Learner Feedback']
```

```python
# TREE FOR: Tutor Feedback + Socially-Oriented Learner Feedback (Uptake)

# PURPLE BOX 47
filtered_df = control_df[control_df['average_normalized_num_eliciting'] <= 0.34]
control_average_uptake = filtered_df['uptake_change'].mean()

filtered_df = tutor_learner_social_df[tutor_learner_social_df['average_normalized_num_eliciting'] <= 0.34]
treatment_average_uptake = filtered_df['uptake_change'].mean()

print('Purple box 47')
print(f'Control ATE: {control_average_uptake}')
print(f'Treatment ATE: {treatment_average_uptake}')
print(f'CATE: {treatment_average_uptake - control_average_uptake}')

# PURPLE BOX 32
filtered_df = control_df[control_df['average_normalized_num_eliciting'] > 0.34]
control_average_uptake = filtered_df['uptake_change'].mean()

filtered_df = tutor_learner_social_df[tutor_learner_social_df['average_normalized_num_eliciting'] > 0.34]
treatment_average_uptake = filtered_df['uptake_change'].mean()

print('Purple box 32')
print(f'Control ATE: {control_average_uptake}')
print(f'Treatment ATE: {treatment_average_uptake}')
print(f'CATE: {treatment_average_uptake - control_average_uptake}')
```
